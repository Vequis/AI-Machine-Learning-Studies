```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
import ipywidgets as widgets
from ipywidgets import interact

n = 4
k = np.arange(0, n + 1)

probabilidades = binom.pmf(k, n, 0.5)

# Função para atualizar o gráfico com base no valor de p
def plot_binomial(p):
    probabilidades = binom.pmf(k, n, p)
    plt.stem(k, probabilidades)
    plt.title(f'Distribuição Binomial (n={n}, p={p:.2f})')
    plt.xlabel('Número de sucessos (k)')
    plt.ylabel('Probabilidade')
    plt.xticks(k)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Criar o slider interativo
interact(plot_binomial, p=widgets.FloatSlider(min=0, max=1, step=0.01, value=0.5))
```

```python
# quero provar na prática que a distribuição é binomial

D = {'N', 'E', 'S', 'W'}

X = np.zeros(len(k))

number_of_samples = 100

for i in range(number_of_samples):
    amostra = np.random.choice(list(D), size=4, replace=True)
    X[np.sum(amostra == 'N') + np.sum(amostra == 'S')] += 1

X = X / number_of_samples

plt.subplot(3, 1, 1)
plt.stem(k, X)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.subplot(3, 1, 2)
plt.stem(k, probabilidades)
plt.title('Distribuição Binomial (n=4, p=0.25)')
plt.xlabel('Número de sucessos (k)')
plt.ylabel('Probabilidade')
plt.xticks(k)
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.subplot(3, 1, 3)
plt.stem(k, abs(X - probabilidades))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, 0.4)


plt.tight_layout()
```

```python
cdf = np.cumsum(probabilidades)

plt.step(k, cdf, where='post', color='b')
plt.title('Função de Distribuição Acumulada (CDF)')
plt.xlabel('Número de sucessos (k)')
plt.ylabel('Probabilidade acumulada')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(k)
plt.show()
```

```python
def plot_alpha(alpha):
    prob = 1 / k**alpha
    prob = prob / np.sum(prob)
    plt.stem(k, prob)
    plt.title(f'Distribuição com alpha={alpha:.2f}')
    plt.xlabel('k')
    plt.ylabel('Probabilidade')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

interact(plot_alpha, alpha=widgets.FloatSlider(min=0, max=2, step=0.05, value=1))
```
