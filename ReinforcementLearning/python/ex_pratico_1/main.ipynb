```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_sample_from_soft_max(z, N=1):
    exponentials = np.exp(z)
    soft_max = exponentials / np.sum(exponentials)

    x = np.arange(1, len(z) + 1)

    return np.random.choice(x, p=soft_max, size=N)
```

```python
z = [1, 1, 1]

print(generate_sample_from_soft_max(z, 3))
```

```python
np.arange(1, 11)
sample = generate_sample_from_soft_max(np.arange(1, 11), 10)

print(sample)
```

```python
counts = np.bincount(sample)[1:] # ignore 0
print(counts)

plt.stem(np.arange(1,11), counts/np.sum(counts))
plt.plot(np.arange(1,11), np.exp(np.arange(1,11))/np.sum(np.exp(np.arange(1,11))), 'r--')
```
